# Data Lakes with Spark

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf, col, monotonically_increasing_id
from pyspark.sql.functions import year, month, dayofmonth, hour, weekofyear, date_format
from pyspark.sql.types import IntegerType, StringType, TimestampType, DateType, DatetimeConverter
from datetime import datetime
```

```python
spark = SparkSession.builder.config("spark.jars.packages", "org.apache.hadoop:hadoop-aws:2.7.0").getOrCreate()
```

### Process Song data

```python
songdf = spark.read.json("s3a://udacity-dend/song_data/*/*/*")
```

```python
songdf.printSchema()
```

```python
songdf.limit(5).toPandas()
```

```python
songdf = songdf.withColumn("year", col("year").cast(IntegerType())) \
    .withColumn("num_songs", col("num_songs").cast(IntegerType()))
```

```python
songdf.printSchema()
```

```python
# Create songs table
songs_table = songdf.filter(col("song_id").isNotNull()) \
    .select("song_id", "title", "artist_id", "year", "duration") \
    .drop_duplicates(["song_id"])
songs_table.limit(5).toPandas()
```

```python
songs_table.limit(5).show()
```

```python
# Save into parquet
songs_table.write.partitionBy("year", "artist_id").parquet("songs_table")
```

```python
# Create artists table
artists_table = songdf.filter(col("artist_id").isNotNull()) \
    .select("artist_id", "artist_name", "artist_location", "artist_latitude", "artist_longitude") \
    .drop_duplicates(["artist_id"])
artists_table.limit(5).toPandas()
```

```python
# Save into parquet
artists_table.write.parquet("artists_table")
```

### Process Log data

```python
logdf = spark.read.json("log-data/*")
```

```python
logdf.printSchema()
```

```python
logdf.limit(5).toPandas()
```

```python
logdf = logdf.withColumn("itemInSession", col("itemInSession").cast(IntegerType())).withColumn("userId", col("userId").cast(IntegerType()))
```

```python
logdf.printSchema()
```

```python
# Create users table
users_table = logdf.filter(col("userId").isNotNull()) \
    .select("userId", "firstName", "lastName", "gender", "level") \
    .drop_duplicates(["userId"])
```

```python
users_table.limit(5).toPandas()
```

```python
users_table.write.parquet("users_table")
```

```python
# Create timestamp column from original timestamp column
@udf(TimestampType())
def to_timestamp(ts):
    timestamp = datetime.fromtimestamp(ts // 1000)
    return timestamp

logdf = logdf.withColumn("timestamp", to_timestamp("ts"))
```

```python
logdf.limit(5).toPandas()
```

```python
# Create datetime column from timestamp column
@udf(DateType())
def to_datetime(ts):
    dt = datetime.fromtimestamp(ts // 1000)
    return dt
logdf = logdf.withColumn("dt", to_datetime("ts"))
```

```python
logdf.limit(5).toPandas()
```

```python
year = udf(lambda x: x.year, IntegerType())
month = udf(lambda x: x.month, IntegerType())
day = udf(lambda x: x.day, IntegerType())
week = udf(lambda x: x.isocalendar()[1], IntegerType())
hour = udf(lambda x: x.hour, IntegerType())
weekday = udf(lambda x: x.weekday(), IntegerType()
```

```python
# Create time table
time_table = logdf.withColumn("year", year("timestamp")) \
    .withColumn("month", month("timestamp")) \
    .withColumn("day", day("timestamp")) \
    .withColumn("week", week("timestamp")) \
    .withColumn("hour", hour("timestamp")) \
    .withColumn("weekday", weekday("timestamp")) \
    .selectExpr("ts AS start_time", "hour", "day", "week", "month", "year", "weekday").drop_duplicates(["start_time"])
```

```python
time_table.limit(5).toPandas()
```

```python
time_table.write.partitionBy("year", "month").parquet("time_table")
```
